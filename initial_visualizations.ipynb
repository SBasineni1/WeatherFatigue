```python
type_counts = events.groupby(['state', 'typetext'])['typetext'].count().unstack()
row_total = type_counts.sum(axis=1)
type_counts_norm = type_counts.div(row_total, axis=0)
type_counts_norm
```

```python
sns.heatmap(type_counts_norm)
```
